```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/Class__/50_Startups.csv'
data = pd.read_csv(file_path)
```

```python
# Setting up the plot style
sns.set(style="whitegrid")
```

```python
# 1. Summary Statistics
# Understand the central tendency, dispersion, and potential anomalies in the data.
summary_stats = data.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
# 2. Distribution Analysis - Histograms and Box Plots
# To visualize the distribution and identify potential outliers for each numerical feature.
for col in data.select_dtypes(include='float').columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f'Distribution and Boxplot of {col}')

    # Histogram
    sns.histplot(data[col], kde=True, ax=axes[0])
    axes[0].set_title(f'{col} - Histogram')

    # Boxplot
    sns.boxplot(x=data[col], ax=axes[1])
    axes[1].set_title(f'{col} - Box Plot')

    plt.show()
```

```python
# 3. Correlation Analysis
# Purpose: To examine relationships between numerical features and identify strongly correlated pairs.
plt.figure(figsize=(8, 6))
# Selecting only numerical columns for correlation analysis
numerical_data = data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# 4. Categorical Analysis - Profit by State
# To analyze the average profit across different states and check for state-wise trends.
plt.figure(figsize=(8, 5))
sns.boxplot(x="State", y="Profit", data=data, palette="Set2")
plt.title("Profit Distribution by State")
plt.show()
```
